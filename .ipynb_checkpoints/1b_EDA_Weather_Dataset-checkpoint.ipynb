# 01. Exploratory Data Analysis

```python
# Import libraries.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Read weather data from csv file.
weather_df = pd.read_csv("./assets/weather.csv")
```

```python
print("######################################")
print("       Info Of the Data Set           ")
print("######################################")
weather_df.info()
```

**Observation:** Date data type is object. It will be converted to datetime later.

```python
# Print number of rows and columns in dataframe.
print(f"Number of rows and columns:", weather_df.shape)

# Display first 5 rows of dataframe.
weather_df.head()
```

## a. Date

```python
# Convert "Date" data type to datetime.
weather_df['Date'] = pd.to_datetime(weather_df['Date'])

print("######################################")
print("       Info Of the Data Set           ")
print("######################################")
weather_df.info()
```

```python
# Convert Date to Yeat, Month, Week & Day for EDA
# Convert to week: https://stackoverflow.com/questions/2600775/how-to-get-week-number-in-python
weather_df['Year'] = weather_df['Date'].dt.year
weather_df['Month'] = weather_df['Date'].dt.month
weather_df['Week'] = weather_df['Date'].dt.isocalendar().week

weather_df.head(3)
```

```python
# Check the timeframe of the data collections
display(weather_df['Date'].min())
display(weather_df['Date'].max())
display(weather_df['Year'].unique())
display(weather_df['Month'].unique())
```

**Observations:**
- Datas are fist collected on 1st May 2007, and last collected on 31st October 2014
- Datas are collected From May - October (6 months) every year from 2007 to 2014

## b. Tavg

```python
weather_df['Tavg'].unique()
```

```python
# Based on the data dictionary, 'M' is missing value. Convert 'M' to None
weather_df = weather_df.replace('M', np.nan)
```

```python
# Convert Tavg & other features to float
weather_df['Tavg'] = weather_df['Tavg'].astype(float)
```

```python
weather_df.info()
```

```python
# Lineplot on Week vs Feature
def lineplot_year(df, feature):

    year_list = [2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]
    
    for year in year_list:
        fig, ax = plt.subplots(figsize = (21, 6))
        sns.lineplot(data = df, x = df[df['Year'] == year]['Week'], y = df[feature], hue = 'Station', palette = 'bright', ci = None)
        plt.xticks(range(15, 48))
        plt.title(f"Plot of {feature} in year {year}")
        plt.show()
```

```python
lineplot_year(weather_df, 'Tavg')
```

**Observations:** Station 2 has higher Tavg as compared to Station 1. As observed, these gap increases with time.

## c. DewPoint

```python
lineplot_year(weather_df, 'DewPoint')
```

**Observations:** Similar to Tavg, Station 2 also has higher Dewpoint as compared to Station 1. As observed, these gap increases with time.

## d. PrecipTotal

```python
weather_df['PrecipTotal'].unique()
```

```python
# Based on the data dictionary, the 'T' represent 'Trace' hence we could replace it as '0'
weather_df.loc[weather_df['PrecipTotal'] == '  T', 'PrecipTotal'] = 0
```

```python
# Convert Tavg to float
weather_df['PrecipTotal'] = weather_df['PrecipTotal'].astype(float)
```

```python
lineplot_year(weather_df, 'PrecipTotal')
```

**Observations:** Station 1 & Station 2 are located considerably near (~24 miles or ~40km). However from the graph above, it can be observed that the both stations have the same rain pattern but the intensity of the rain are quite different.

## e. WetBulb

```python
lineplot_year(weather_df, 'WetBulb')
```

**Observations:** Similar to Tavg & DewPoint, Station 2 also has higher WetBulb as compared to Station 1. As observed, these gap increases with time.

## f. Feature Analysis

**(i) Univariate Analysis**

This is to analyse each individual variable or feature. This univariate analysis describes how data in each feature is distributed.

```python
# Rename columns.
weather_df.rename(columns = {'AddressAccuracy':'GeoCoder Accuracy', 
                           'NumMosquitos':'Number Mosquitoes Caught',
                           'WnvPresent': "West Nile Virus in Mosquitos" 
                           }, inplace = True)
weather_df.columns
```

```python
# Draw histogram.
ax = weather_df.hist(bins=25, grid=False, figsize=(13,13), color='skyblue', zorder=2, rwidth=0.9);
```

```python
def boxplot_distplot(dataframe, number_of_rows, number_of_columns, data_type):

    # create a list to store object type 'int64' and 'float64'.
    columns_list = data_type
    
    # create an object to store numerical columns
    all_columns = dataframe.select_dtypes(include=columns_list).columns

    fig, ax = plt.subplots(number_of_rows, number_of_columns, figsize = (15, 13))

    for column_index in range(len(all_columns)):
        sns.boxplot(data = dataframe, x = all_columns[column_index], ax = ax[column_index,0])
        sns.distplot(dataframe[all_columns[column_index]], ax = ax[column_index,1])

    plt.tight_layout()
```

```python
# silence warning
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

boxplot_distplot(dataframe = weather_df[['Tmax', 'Tmin', 'Tavg', 'DewPoint', 'ResultSpeed', 'ResultDir']], number_of_rows = 6 , number_of_columns = 2 , data_type = ["float64", "int64"])
```

**Observations:**
- there are outliers in both Tmax and ResultSpeed; and
- Tmax, Tmin, DewPoint skew to the left and ResultSpeed skew to the right.

**(ii) Multi-Variate Analysis**

This is to analyse at least two variables or features together and see how they are related to each other.

```python
sns.pairplot(weather_df, diag_kind='kde');   
# to plot density curve instead of the histogram on the diagram # Kernel density estimation(kde)
```

**Observation:**
-  `DewPoint` and `Tmax`; `DewPoint` and `Tmin`; `Tmax` and `Tmin` have linear relationship.

**(iii) Correlation Heatmap**

```python
# silence warning
from warnings import filterwarnings
filterwarnings(action='ignore', category=DeprecationWarning, message='`np.bool` is a deprecated alias')
```

```python
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(weather_df.corr(), dtype=np.bool))

plt.figure(figsize=(10,6))
sns.heatmap(weather_df.corr(),cmap=plt.cm.Blues,annot=True, mask=mask, vmin=-1, vmax=1)
plt.title('Heatmap displaying the relationship between the features of the data',
         fontsize=13)
plt.show()
```

**Observations:**
- `DewPoint` and `Tmax`; `DewPoint` and `Tmin`; `Tmax` and `Tmin` have strong positive correlation with each other.
