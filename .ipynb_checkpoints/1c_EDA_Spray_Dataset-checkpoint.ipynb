```python
# Calculate the distance between two points and return 'Y' if within 3km
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles
    if c * r < radius:
        return 'Y'

# center_point = [{'lat': -7.7940023, 'lng': 110.3656535}]
# test_point = [{'lat': -7.79457, 'lng': 110.36563}]

# lat1 = center_point[0]['lat']
# lon1 = center_point[0]['lng']
# lat2 = test_point[0]['lat']
# lon2 = test_point[0]['lng']

radius = 3.00 # in kilometer

a = haversine(lon1, lat1, lon2, lat2)

# print('Distance (km) : ', a)
# if a <= radius:
#     print('Inside the area')
# else:
#     print('Outside the area')
```

## Spray data with Nearby Traps
As per CDC the [lifecycle](https://www.cdc.gov/mosquitoes/about/life-cycles/culex.html) of a mosquito (Culex pipiens, Cx. quinquefasciatus, and Cx. tarsalis) from an egg to an adult takes approximately 7-10 days. The effect of fogging will last around [30 days](https://invatechitalia.com/blogs/news/everything-you-need-to-know-about-a-mosquito-fogger). Thus, we will look into the number of mosquitos 7 days before the spray and 30 days after the after the spray.
From the same CDC article, it mentioned that Culex mosquitos do not fly long distances which is up to 3.2km. In our calculation we take the radius of 3km from the spray.

```python
# Extracting out train data that were between the dates '2011-09-01' to '2011-10-07'
train20110907 = train[(train['Date'] >='2011-09-01') & (train['Date'] <= '2011-10-07')]
train20110907.reset_index(inplace=True,drop=True)
train20110907['within3km'] = ''
```

```python
train20110907
```

```python
# Finding traps that are within 3km of the spray
for i in tqdm(range(len(train20110907)), desc = 'tqdm() Progress Bar'):
    for index in range(len(spray20110907)):
        train20110907['within3km'][i] = haversine(spray20110907['Longitude'][index], spray20110907['Latitude'][index], train20110907['Longitude'][i], train20110907['Latitude'][i])
        index += 1
    i += 1
```

```python
# Traps that are within 3km of the spray region
train20110907[train20110907['within3km'] == 'Y']
```
